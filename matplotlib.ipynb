```python
# red, 별모양, dash선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기
theta=np.linspace(0,2*np.pi,20)
plt.plot(theta,np.sin(theta),"r*--",markersize=10, label="sin")
# plt.figure() # 새창 열기. 이거 없으면 겹쳐 그리기
plt.plot(theta,np.cos(theta),"go-",markersize=10, label="cos")
plt.xlabel("theta")
plt.ylabel("value")
plt.title("sin and cos",color="b")
plt.legend()
plt.grid()
# plt.xlim([0, np.pi])
# plt.ylim([-2, 2])
plt.axis([0, np.pi, -2,2])
```

```python
plt.plot(theta,np.sin(theta),"r*--",markersize=10, label="sin")
plt.plot(theta,np.cos(theta),"go-",markersize=10, label="cos")
plt.xlabel("theta",fontsize=20)
plt.ylabel("value",fontsize=20)
plt.title("sin and cos",color="b",fontsize=20)
plt.legend(fontsize=20)
plt.grid()
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

## plt.bar() 와 plotyy

```python
plt.bar(range(5),[3,2,3,4,5], width=0.8)
plt.xticks(range(5),["a","b","c","d","e"])
plt.xlabel("layer")
plt.ylabel("value")
```

```python
x=range(5)
y1=[i for i in range(1,6)]
y2=[i+100 for i in range(1,6)]

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1=ax1.plot(x,y1,'b')
p2=ax2.plot(x,y2,'r')
ax1.set_ylim([-10, 10])
ax1.set_xlabel("x")
ax1.set_ylabel("y1")
ax2.set_ylabel("y2")
ax1.set_title("y1 and y2")
ax1.grid()
plt.legend(p1+p2,["y1","y2"], loc="right")
```

## 2개 변수 함수: 등고선으로 보기, 3D 로 보기



```python
a=np.linspace(-1,1,100)
b=np.linspace(-1,1,100)
A, B=np.meshgrid(a,b)
L = A**2 + B**2
plt.contour(a,b,L,20) # 20 -> 등고선 개수
plt.xlabel('a'); plt.ylabel('b'); plt.grid()

plt.figure()
plt.contourf(a,b,L,5)
```

```python
fig, ax = plt.subplots(figsize=[10, 9])
CS = ax.contour(a,b,L,20) # 20 -> 등고선 개수
ax.clabel(CS, fontsize=10)
plt.xlabel('a'); plt.ylabel('b'); plt.grid()
```

```python
plt.figure(figsize=[10, 9]) # [가로, 세로]
ax = plt.axes(projection="3d")
ax.plot_surface(A,B,L)

# 이전 코드 (error)
# figure = plt.figure(figsize=[10, 9])
# ax = figure.gca(projection="3d")
# ax.plot_surface(A,B,L)

plt.figure(figsize=[10, 9])
ax = plt.axes(projection='3d')
ax.plot_wireframe(A,B,L)
```

```python
plt.figure(figsize=[10,9]) # [가로, 세로]
ax = plt.axes(projection='3d')
ax.view_init(elev=10,azim=-45)
ax.plot_surface(A,B,L, alpha=0.7)

plt.figure(figsize=[10, 9])
ax = plt.axes(projection='3d')
ax.view_init(elev=10,azim=-45)
ax.plot_wireframe(A,B,L)
```

## plotly

```python
import plotly.graph_objects as go

a=np.linspace(-2,2,100)
b=np.linspace(-2,2,100)
A, B = np.meshgrid(a,b)
L = np.cos(A**2.4)+np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title='cos(A^2.4)+sin(B^2.4)', width=700, height=600)
```

```python
fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title='cos(A^2.4)+sin(B^2.4)', width=700, height=600,
                  scene = dict( zaxis=dict(nticks=40, range=[-3,3]) ))
```

## 이미지 보기

```python
import matplotlib.pyplot as plt
from torchvision import datasets
test_DS = datasets.MNIST(root="data/", train=False, download=True) # 폴더 한번 확인 해보자
```

```python
print(test_DS.data.shape)
print(test_DS.data[0])
```

```python
plt.imshow(test_DS.data[0]) # image show
```

```python
plt.imshow(test_DS.data[3], cmap="gray")
plt.colorbar()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=[8,5]) # object-oriented
print(type(fig))
print(type(ax))
for i in range(2):
    for j in range(3):
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
        ax[i,j].imshow(test_DS.data[3*i+j], cmap='gray')
        ax[i,j].set_title(test_DS.targets[3*i+j].item(),color='g')
```

```python
im=test_DS.data[0]
plt.figure(figsize=(20,15))
plt.imshow(im, cmap="gray")
plt.colorbar()

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        plt.text(j,i, im[i,j].item(),
                 horizontalalignment="center",
                 color="k" if im[i,j] > torch.max(im) / 1.5 else "w")
plt.xticks([])
plt.yticks([])
```
